```python
# Read the data from files
chunk_size = 10000
# Read data from Signal Data
file_path_2Gb_signal = '/content/drive/MyDrive/2Gb_signal.csv'
signal_data = pd.read_csv(file_path_2Gb_signal, chunksize=chunk_size)
# Read data from suction
file_path_suction = '/content/drive/MyDrive/Only_re_suction_peaks.csv'
suctions_data = pd.read_csv(file_path_suction, chunksize=chunk_size)
# Read water data
file_path_just_water = '/content/drive/MyDrive/just_water.csv'
water_data = pd.read_csv(file_path_just_water, chunksize=chunk_size)
# print(signal_data)
# print(suctions_data)
# print(water_data)
```

```python
# Process the 2G Signal Data
for signal in signal_data:
     # Calculate the mean of 'adc1' column
    adc1_mean = signal['adc1'].mean()

    # Subtract the mean to center the data around zero
    centered_adc1 = signal['adc1'] - adc1_mean

    # Plot the centered 'adc1' column
    plt.figure(figsize=(12, 6))
    plt.plot(centered_adc1, label='adc1 (centered)', color='red')

    # add labels, legend
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.title('Marginum Data-2G Signal Data')
    plt.grid(True)
    # Show the plot
    plt.show()

    # Break the loop after processing the first chunk
    break
```

```python
# Process suction data
for suction in suctions_data:
    # Calculate the mean of 'adc1' column
      adc1_mean = suction['adc1'].mean()
    # Subtract the mean to center the data around zero
      centered_adc1 = suction['adc1'] - adc1_mean
    # Plot the centered 'adc1' column
      plt.figure(figsize=(12, 6))
      plt.plot(centered_adc1, label='adc1 (centered)', color='black')
    # Plot the 'adc1' and 'adc2' columns
      # print(water)
      plt.xlabel('Index')
      plt.ylabel('Value')
      plt.legend()
      plt.title('Marginum suction Data')
      plt.grid(True)
      # Show the plot
      plt.show()
      break
```

```python
# Process water data
for water in water_data:
    # Calculate the mean of 'adc1' column
      adc1_mean = water['adc1'].mean()
    # Subtract the mean to center the data around zero
      centered_adc1 = water['adc1'] - adc1_mean
    # Plot the centered 'adc1' column
      plt.figure(figsize=(12, 6))
      plt.plot(centered_adc1, label='adc1 (centered)', color='blue')
    # Plot the 'adc1' and 'adc2' columns
      plt.xlabel('Index')
      plt.ylabel('Value')
      plt.legend()
      plt.title('Marginum Data-2G Signal Data')
      plt.grid(True)
      # Show the plot
      plt.show()
      break
```
